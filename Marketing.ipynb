## 14. Perform Correlation Analysis and write your key inferences

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap = 'coolwarm')
```

Based on the given data, here are some key inferences:

1. Income is positively correlated with various types of purchases such as Wine, Meat Products, Fish Products, Fruits, and Gold Products. This suggests that customers with higher incomes are more likely to spend more on these items. This information could be valuable for marketing and targeting high-income customers for specific product categories.

2. The number of Web Visits per Month has a negative correlation with Income. Customers with higher incomes tend to visit the website less frequently. This indicates that online marketing strategies or website features may need to be adjusted to attract and engage higher-income customers.

3. Age (Year_Birth) has a moderate negative correlation with Income. Older customers tend to have lower incomes. This highlights the income disparity among different age groups and can help inform marketing campaigns or product offerings targeted towards specific age demographics.

4. The Response variable has positive correlations with Accepted Campaigns. Customers who responded positively to previous campaigns are more likely to respond to the current campaign. This suggests that successful marketing campaigns can have a cumulative effect on customer engagement and response rates.

5. Kidhome (number of children at home) has a negative correlation with Income. Customers with more children tend to have lower incomes. This information can be useful for developing marketing strategies or product bundles tailored to the needs and budgets of customers with children.

6. Complain has a very weak positive correlation with Income. This suggests that customers with higher incomes may be slightly more prone to lodging complaints. Monitoring and addressing customer complaints, especially among higher-income customers, can be crucial for maintaining customer satisfaction and loyalty.
